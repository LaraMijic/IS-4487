## Summarize Data

➡️ Assignment Tasks
- Display the average price per month
- Display the total revenue for all months

Average Price Per Month

```python
average_price_by_month = df_cleaned.groupby('order_month')['product_price'].mean()

print(average_price_by_month)
```

Total Revenue

```python
revenue = df_cleaned['product_price'] * df_cleaned['quantity']
total_revenue = revenue.sum()
print(total_revenue)
```

Total Revenue by Month

```python
df_cleaned['total_revenue'] = revenue

# Group by 'order_month' and calculate the sum of total_revenue for each month
monthly_revenue = df_cleaned.groupby('order_month')['total_revenue'].sum()

print(monthly_revenue)
```
