## Clean up the data
Did you notice anything unusual about the "house" amounts?

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Summarize the Data

## Data Visualization

 Assignment Tasks
- Add at least two charts to show a summary of any major trends

### Setup for Charts:

```python
numerical_predictors = ['income', 'overage', 'leftover', 'house', 'handset_price', 'over_15mins_calls_per_month', 'average_call_duration']

categorical_predictors = ['college', 'reported_satisfaction', 'reported_usage_level', 'considering_change_of_plan', 'leave']
```

### Chart Type 1: Boxplots

```python
for predictor in numerical_predictors:
    sns.boxplot(x='leave', y=predictor, data=df_clean)
    plt.show()
```

### Chart Type 2: Stacked Barcharts



```python
for predictor in categorical_predictors:
    pd.crosstab(df_clean[predictor], df_clean['leave']).plot(kind='bar', stacked=True)
    plt.show()
```
